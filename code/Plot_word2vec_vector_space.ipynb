# Plot word2vec vector spaces

We use dimentionality reduction to plot the 300 dimensions word2vec vector space in 2D. Plotting takes a bit of computer power, because we try to avoid having overlapping labels. 

```python
# general:
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline


import time # will be use to choose the faster solution
from sklearn.decomposition import PCA
from adjustText import adjust_text # to avoid overlapping text annotations on plots
import pickle # to load previously saved custom models
```

```python
# load model:
pkl_filename = "../data/custom_wine_word2vec_model.pkl"
with open(pkl_filename, 'rb') as file:
    custom_wine_word2vec_model = pickle.load(file)
```

```python
# apply PCA 
X = custom_wine_word2vec_model[custom_wine_word2vec_model.wv.vocab]
pca = PCA(n_components=2)
result = pca.fit_transform(X)
```

**The variance explained by the 2 components is extremely low**. The 300 dimensions of the woird2vec model are not easily reduced. Thus it is likely that the plots of the features in 2D will look quite random and be useless.

```python
pca.explained_variance_
pca.explained_variance_ratio_
```

```python
# plot the vocabulary of the trained model in 2D.
# takes time to run because of "adjust_text"

time0 = time.time()

plt.figure(figsize=(16, 16))
p = sns.scatterplot(result[:, 0], result[:, 1], color='skyblue')
# add annotations one by one with a loop
texts=[]
for line in range(0,result.shape[0]):
    if line % 100 == 0: # annotate only some words
        texts.append(plt.text(result[line, 0], result[line, 1], 
                              list(custom_wine_word2vec_model.wv.vocab)[line], 
                              horizontalalignment='center', size='large', 
                              color='black', weight='semibold'))
# adjust_text(texts, autoalign='xy', expand_objects=(0.1, 1),
#             only_move={'points':'', 'text':'y', 'objects':'y'}, force_text=0.75, force_objects=0.1)
adjust_text(texts)
plt.savefig("../figures/custom_word2vec_vec_spave_pca.png", bbox_inches='tight', dpi=300)
plt.show()

print("plotted in", time.time()-time0, "seconds")
```

Let's try to reduce the number of features that we plot, by selecting randomly 100 features.  

```python
# comp_1 = result[:, 0]
# comp_2 = result[:, 1]
comps = { 'comp_1': result[:, 0], 'comp_2': result[:, 1] } 
vocab = list(custom_wine_word2vec_model.wv.vocab.keys())
two2_space = pd.DataFrame(comps, index=vocab)

# we sample 100 of the 17986 features:
two2_space_samples = two2_space.sample(n=100, random_state=0)

plt.figure(figsize=(16, 16))
sns.scatterplot(two2_space_samples.comp_1, two2_space_samples.comp_2, color='steelblue')

# add annotations one by one with a loop
texts=[]
for line in range(0,two2_space_samples.shape[0]):
    texts.append(plt.text(two2_space_samples.comp_1[line],
                          two2_space_samples.comp_2[line], 
                          list(two2_space_samples.index)[line], 
                          horizontalalignment='center', 
                          size='large', 
                          color='black', 
                          weight='semibold'))
adjust_text(texts)
plt.savefig("../figures/custom_word2vec_vec_spave_pca_100.png", bbox_inches='tight', dpi=300)
plt.show()

```
